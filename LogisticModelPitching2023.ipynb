```python
#Dependcies 

from sklearn import *
from sklearn.model_selection import GroupKFold
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.model_selection import KFold
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
```

```python
#Read The File 

df=pd.read_csv("Log20222023MLBPitching.csv")
df
```

```python
# Read The Columns 

df.columns
```

```python
# Lets Try Again With Logistic Regression and GNB For Pitching To Determine Who is 
# The Best team to go to the playoffs

inputs=df[['R','H','SO']]
target=df[["PLAYOFFS"]]

X_train,X_test,y_train,y_test=train_test_split(inputs,target,train_size=0.8)
LogReg=LogisticRegression()
GNB=GaussianNB()

# KFolds Cross-Validation 
folds=StratifiedKFold(n_splits=10)
```

```python
# Fit The Logisitic Model 
model=LogReg.fit(X_train,y_train)
model2=GNB.fit(X_train,y_train)

# Model With KFolds Score
print("LogReg CrossValidaton Score: ",cross_val_score(LogisticRegression(), inputs,target))
print("GNB CrossValidaton Score: ",cross_val_score(GaussianNB(), inputs,target))
```

```python
# Predict the values 
y_pred=model.predict(X_test)
y_pred2=model2.predict(X_test)

```

```python
# Accuracy Score 
print(accuracy_score(y_pred,y_test))
print(accuracy_score(y_pred2,y_test))
```
